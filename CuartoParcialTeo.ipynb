## (0.5 Puntos) Bases

*   ¿Qué diferencia hay entre una colección de datos y una base de datos?.
1. Una colección de datos tiene información desorganizada que puede estar en cualquier formato o estructura. Sirve para almacenar informaicón pero no es nada facil trabajar con ellos o tratar de obtener información especifica.
2. Una base de datos: Son datos organizados sistematicamnete para que sea más facil el acceso, la gestion. Tiene que estar estructurada de acuerdo a algún modelo que puede resr relacional para que se pueda obtener la ifnormaicón deseada. Y se utiliza algún  gestor para esto como puede ser MySQL.

*   ¿Qué diferencia hay entre *SQL* y *MySQL*?
SQL es el lenguaje, mientras que MySQL es un software que usa ese lenguaje para gestionar bases de datos.

# (0.5 Puntos) Estándares (*XML, JSON*)

## (0.5 Puntos) *XML* y *JSON*

*   ¿Para que sirven estándares como *XML* o *JSON*?.

Sirven para estructurar, alamcenar e intercambiar datos de manera eficiente entre sistemas. Facilitan  el uso de informacion organizandolo de forma jerarquica. Son de uso univer

*    ¿Cuál es la principal diferencia entre *XML* y *JSON*?.

La principal diferencia entre XML y JSON es su forma de estructurar los datos: XML usa etiquetas anidadas similares al HTML, lo que lo hace más verboso y pesado, mientras que JSON organiza los datos en pares clave-valor y estructuras como listas y objetos, siendo más compacto, fácil de leer y eficiente. JSON es ideal para APIs modernas, mientras que XML se usa en sistemas que requieren validación estricta o datos más complejos.
